## 3.3 순서인수, 키워드인수

- 키워드인수 : 인수의 이름을 명시하는 방식

````python
def func(a, L=None):
    if L is None: # None을 비교할때는 ==가 아닌 is를 사용해야함.
        L = [ ]
    L.append(a)
    return L
````

````python
lst = [ ]
print(func(10, lst)) # 순서인수
print(func(L=[ ], a=10)) # 키워드인수
print(func(20)) # 순서인수 a에 20이 전달
print(func(20, None))
````

````python
# args앞의 * : 매개변수들(인수들)이 args 튜플로 전달
def add(*args):
        "전달받은 매개변수들의 합을 return. 매개변수가 없을 경우 0을 return"
        #print(args) # ()
        tot = 0 # 누적변수
        for num in args:
            tot += num # tot = tot + num
        return tot
````

## 3.7 인수의 언패킹

- 패킹 : 함수 정의시 매개변수 앞에 *붙임(튜플매개변수)

````python
# ex5. safe_index(ch03.py)를 람다식으로
def safe_index(vector, item):
    '''
    첫번째 매개변수 lst에서 item 요소가 있는 index를 반환하는 함수
    item 요소가 없으면 -1를 반환
    '''
    if item in vector:
        return vector.index(item)
    return '-1 없음'
````

## 4.2 함수의 매개변수(인수)에 람다식 사용
- 람다식을 사용하는 곳 : 함수의 매개변수 > 함수를 return
### (1) map 

````python
def map_template(func, L=[ ]):
    result = [ ]
    for item in L:
        result.append(func(item))
    return result
````

````python
lst = [1, 2, 3, 4, 5, 6, 7, 8]
map_template(lambda x:x*2, lst)
````
