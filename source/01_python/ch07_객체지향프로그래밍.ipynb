# 2절. 생성자(`__init__(self, ...)')와 소멸자(`__del__(self)`)
```
생성자 : __init__(self, 매개변수, ..) 객체가 생성될 때 자동 실행
            생성시 필요한 코드(인스턴스 변수 초기화)를 포함
            없으면 기본(default) 생성자 자동 생성
소멸자 : __del__(self) 객체가 소멸될(참조한 변수가 없어질) 때 자동 실행
            인스턴스 객체의 래퍼런스 카운터가 0이 될 때 실행
            소멸시 필요한 코드를 포함
```

````python
class Person:
    def __init__(self, name, gender):
            self.name = name
            self.gender = gender
            print('Person 생성자 호출(객체 생성 == 클래스형 변수 생성 == 인스턴스 메모리 생성)')
    def __del__(self):
        print('Person 객체 소멸')
    def print_info(self):
           print(f"{self.name}님은 {self.gender}입니다")
    def __str__(self):
           return f"{self.name}은 {self.gender}이죠"
````

````python
p1 = Person('내이름은 코난', '여자')
p1.print_info()
print(p1) 
# print(p1) = print(p1.__str__())
````

````python
type(p1), isinstance(p1, Person)
````

````python
del p2 # 레퍼런스 카운트가 0이 아님
````

````python
del p3 # 레퍼런스 카운트가 0
````

````python
# 일반함수의 중복정의 불가
def add(a,b):
       return a+b
def add(a,b,c):
    return a+b+c
add(1, 2, 3)
````

````python
class Person:
    # 생성자를 이용한 인스턴스 변수(멤버변수) 초기화
#     def __init__(self):
#         self.name = "아무개"
#         self.gender = "남자"
#     def __init__(self, name):
#         self.name = name
#         self.gender = "남자"
# 생성자 함수 중복 저의 불가. 매개변수의 기본값을 갖는 매개변수를 이용한 생성자 정의
    def __init__(self, name="아무개", gender="남자"):
        self.name = name
        self.gender = gender
    def __str__(self):
        return f"{self.name}는(은) {self.gender}이네요"
````

````python
p1 = Person()
p2 = Person("홍길동")
p3 = Person("신길숙", "여자")
print('p1 =', p1)
print('p2 =', p2)
print('p3 =', p3)
````

# 3절. 변수와 메소드(함수)
- 인스턴스 변수(멤버변수) vs 클래스변수 vs static변수(정적변수)
- 인스턴스메소드(self) vs 클래스메소드(cls) vs static메소드()

````python
# 클래스변수 : 클래스 내부 선언 => 생성된 모든 인스턴스에 동일한 값을 가짐
# 클래스.클래스변수를 수정하면 모든 인스턴스에 반영될 수 있음
# 객체변수.클래스변수를 수정하면 해당 객체에만 반영
class Person:
    academy = "MBC"
    def __init__(self, name, gender): # self: 내 객체 자신
        self.name = name
        self.gender = gender
        print('Person 객체 생성합니다')
    def __str__(self):
           return r"%s의 %s은(는) %s입니다" % (self.academy, self.name,self.gender)
print(Person)
print(Person.academy)        
# 0707 내일부터 static 변수 : 클래스 내부 선언. static변수 앞에 under score 2개(__)를 붙임
````

````python
print(p1)
print(p2)
````

````python
Person.academy = "KBS"
print(p1)
print(p2)
p2.academy = "딴데"
p2.name = "홍길동"
print(p1)
print(p2)
````
