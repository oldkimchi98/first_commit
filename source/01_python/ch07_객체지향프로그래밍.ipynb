# 2절. 생성자
(`__init__(self, ...)')와 소멸자(`__del__(self)`)
```
생성자 : __init__(self, 매개변수, ..) 객체가 생성될 때 자동 실행
            생성시 필요한 코드(인스턴스 변수 초기화)를 포함
            없으면 기본(default) 생성자 자동 생성
소멸자 : __del__(self) 객체가 소멸될(참조한 변수가 없어질) 때 자동 실행
            인스턴스 객체의 래퍼런스 카운터가 0이 될 때 실행
            소멸시 필요한 코드를 포함
```

````python
class Person:
    def __init__(self, name, gender):
            self.name = name
            self.gender = gender
            print('Person 생성자 호출(객체 생성 == 클래스형 변수 생성 == 인스턴스 메모리 생성)')
    def __del__(self):
        print('Person 객체 소멸')
    def print_info(self):
           print(f"{self.name}님은 {self.gender}입니다")
    def __str__(self):
           return f"{self.name}은 {self.gender}이죠"
````

※ class 내의 메소드(함수) 종류

- 인스턴스메소드(일반함수) : 첫번째 매개변수에 self를 꼭 명시. 객체를 통해서만 호출할 메소드
- 클래스메소드 : 메소드위에 @classmethod이 선언된 메소드. 첫번째 매개변수에 cls(clz)를 꼭 명시
- static메소드 : 메소드위에 @staticmethod이 선언된 메소드. 디폴트 매개변수가 없음

````python
class Person:
    academy = 'MBC' # 클래스변수
    def __init__(self, name='홍길동', gender='남자'): # 생성자
        self.name = name
        self.gender = gender
    def print_info(self): #인스턴스메소드 : 변수.print_info()로 호출
        print(f'인스턴스메소드(일반함수) : {self.academy}의 {self.name}님은 {self.gender}입니다')
    def change_info(self, name, gender='남자'): # 인스턴스메소드 : 변수.change_info("이름","성별")
        self.name = name
        self.gender = gender
        print(f'인스턴스메소드(일반함수)를 통해 이름은 {self.name}으로 성별은 {self.gender}로 변경')
    @classmethod
    def change_class_method(clz, academy): #클래스메소드 Person.change_class_method("KBS")
        # clz == Person
        clz.academy = academy # 클래스 영역의 클래스 변수 수정
        print(f'클래스메소드를 통해 클래스변수 academy를 {academy}로 수정')
    @staticmethod
    def print_static_method(): # static메소드는 클래스변수를 이용하는 경우보다 로직이 우선일 경우
        print(f'static 메소드 : {Person.academy}로 유지')
````

````python
p1 = Person()
p1.print_info()
````

````python
Person.change_class_method('KBS')
Person.print_static_method()
p1.print_info()
````

## super()이용
- super() : "부모 클래스의"를 의미

````python
class Person:
    def __init__(self, name, gender):
        print("Person 객체 생성")
        self.name = name
        self.gender = gender
    def __del__(self):
        print('Person 객체 소멸')
    def print_info(self):
        print("{}님은 {}".format(self.name, self.gender))
    def __str__(self):
        return "{}님은 {}".format(self.name, self.gender)
class Student(Person):
    def __init__(self, name, gender, major):
          # Person.__init__(self, name, gender)  
        super().__init__(name, gender)  
        self.major = major
    def __del__(self):
          pass
    def print_info(self):
            super().print_info()
            print('전공은 {}입니다'.format(self.major))
    def __str__(self):  # 재정의
          return super().__str__() + "이고, 전공은 {}입니다".format(self.major)
````

## static변수(정적변수) : 여러 객체들 사이에 데이터를 공유하고 싶은 때 사용

*static 변수앞에는 under score 2개(__)를 붙이면, 클래스명._클래스명_static변수명이름으로 참조
* 클래스변수는 클래스나 인스턴스나 외부에서 바로 접근 가능. 반면 __로 시작하는 변수는 외부에서 클래스명._클래스명_static 변수명과
* 같이 접근해야 하드로 외부에서 실수로 접근하거나 변경하는 것을 방지
- static

````python
class Person:
    def __init__(self, name, gender):
        print("Person 객체 생성")
        self.name = name
        self.gender = gender
    def __del__(self):
        print('Person 객체 소멸')
    def print_info(self):
        print("{}님은 {}".format(self.name, self.gender))
    def __str__(self):
        return "{}님은 {}".format(self.name, self.gender)
class Student(Person):
    __count = 0 # static변수(정적변수, 공유변수) student형 변수 몇 개 생성되었는지 저장
    def __init__(self, name, gender, major):
          # Person.__init__(self, name, gender)  
        super().__init__(name, gender)  
        self.major = major
        Student._Student__count += 1 #객체가 생성될 때마다 __count 1 증가
    def __del__(self):
          Student._Student__count -= 1 # 객체가 소멸될 때마다 __count 1 감소
    def print_info(self):
            super().print_info()
            print('전공은 {}입니다 -이상 {}개 생성'.format(self.major, Student._Student__count))
    def __str__(self):  # 재정의
          return super().__str__() + "이고, 전공은 {}입니다 -이상 {}개 생성".format(self.major, Student._Student__count)
        
    @classmethod
    def get_count(cls):
           # return Student._Student__count
            return cls.__count
````

````python
s1 = Student("홍", "남자", "AI")
print(s1)
print(s1.get_count())
print(Student.get_count())
````

````python
Student.__count = 0 # 외부에서 static변수 실수로 수정못하도록 만드는 용도
Student.get_count()
````

````python
# 문제 1.
class Shape:
    def __init__(self, x=0, y=0):
        self.x = x
        self.y = y
    def move(self, x, y):
        self.x = x
        self.y = y
    def __str__(self):
        return "x좌표 : {} , y좌표 : {}".format(self.x, self.y)
    @staticmethod
    def staticmethod_():
        print('Shape클래스의 static 메소드 호출합니다')
        
class Triangle(Shape):
    __count = 0 # static변수(정적변수, 공유변수)
    def __init__(self, width, height, x=0, y=0):
        super().__init__(x, y)
        self.width  = width
        self.height = height
        Triangle._Triangle__count += 1 # 객체가 생성될 때마다 __count 1증가
    def __del__(self):
        Triangle._Triangle__count -= 1 # 객체가 소멸될 때마다 __count 1감소
    def calc_area(self):
        return self.width * self.height * 0.5
    def __str__(self): # 재정의
        return super().__str__() + ", width : {} height : {}, area : {}".format(self.width, 
                                                                                self.height, 
                                                                                self.calc_area())
    @classmethod
    def get_count(cls):
        # return Triangle._Triangle__count
        return cls.__count
````

````python
t1 = Triangle(4, 3, 0, 10)
t2 = Triangle(3, 4, 5, 5)
print(t1)
print(t2)
print('Triangle 객체 갯수 :', Triangle.get_count())
t1.move(50,50)
print(t1)
print(Triangle.get_count())
````

````python
# 8 (3)
class Super:
    def do_(self, a):
        print("super.do_")
class Sub(Super):
    def do_(self, a):
        super().do_(a)
        print("sub.do_")
````

````python
# 9 (2)
class Person:
    '연습문제 9번 예제 만듦'
    def __init__(self, name="hong", gender="M"): # ④생성자
        self.name = name
        self.gender = gender
    def __str__(self): # ② 반드시 메소드 내 self매개변수 필요
        return "name : "+self.name
    def print_info(self):
        print(self.name)
````
