## 1.4 인덱싱과 요소 인덱스 찾기
- 변수[idx] : 인덱스를 이용한 접근(없는 인덱스 사용시 에러)
- count(찾을데이터) : 해당 찾을 데이터가 나오는 갯수 반환 (*인덱스 입력 전에 카운터 먼저 찾아볼것)
- index(찾을데이터) : 0번째부터 해당 찾을데이터가 몇번째 index에 있는지 반환
- index(찾을데이터, start) : start번째부터 해당 찾을데이터가 몇번째 index에 있는지 반환

```python
numbers = [1, 2, 3, [1, 2, 3], 1, 2, 3]
```

```python
numbers.count(2)
```

```python
numbers.count(9)
```

```python
print(2 in numbers) # 2가 numbers의 요소인지 여부
print(9 in numbers)
```

```python
if numbers.count(9):
    print(numbers.index(9))
else:
    print('찾을 수 없음')
```

```python
if 9 in numbers:   # numbers에 9가 있니?
    print(numbers.index(9))
else:
    print('찾을 수 없음')   # 부정문이라면 '찾을 수 없음' 출력.
```

```python
def safe_index(lst, item):
    '''
    첫번째 매개변수 lst에서 item 요소가 있는 index를 반환하는 함수
    item 요소가 없으면 -1를 반환
    '''
    if item in lst:
        return lst.index(item)
    return '없음'
```

```python
from ch03 import safe_index # ch03모듈안의 safe_index 로드
numbers = [1, 2, 3]
safe_index(numbers, 9)
```

## 1.5 원하는 요소 추출

1. 리스트 슬라이싱 : list에서 원하는 index의 데이터들을 추출(인덱스 범위를 벗어나도 에러나지 않음)
    - [from, stop, step] : from부터 stop앞까지 step씩 전진하면서 index 값 추출

## 1.7 요소 삭제하기

```python
# pop()     :  ()는 가장 마지막 인덱스의 요소를 변환하고 삭제
# pop(idx)  :  (idx)번째 요소를 반환하고 삭제
# del 리스트변수[idx]  : [idx]번째 요소를 삭제
# remover(data)   : list안의 data를 삭제 (data가 요소로 없으면 에러)
```

```python
numbers = [1,2,3,4,5]
print('삭제한 데이터 :', numbers.pop())
print('삭제후 numbers :', numbers)
print('삭제한 데이터 :', numbers.pop())
print('삭제후 numbers :', numbers)
print('삭제한 데이터 :', numbers.pop(0))
print('삭제후 numbers :', numbers)
del (numbers[1])
print('삭제후 numbers :', numbers)
```

```python
if 2 in numbers:
    numbers.remove(2)
else:
    print('해당 데이터는 없는 데이터입니다')
```

```python
def safe_remove(lst, remove_item):
    '''
    첫번째 매개변수 lst에서 remove_item 데이터 제거
    해당 remove_item이 lst에 없으면 메세지 출력
    '''
    if remove_item in lst:
        lst.remove(remove_item)
        print('삭제 확인', lst)
    else:
        print('Error - 해당 데이터는 없는 데이터입니다')
```

## 1.9 리스트 복제

```python
# 스칼라 변수의 복제
i = 10
copied_i = i
print(f"원본i={i}, 복수본i={copied_i}) / 원본주소 : {id(i)}, 복사본주소 : {id(copied_i)}")
copied_i += 5 # copied_i = copied_i + 5
print(f"원본i={i}, 복수본i={copied_i}) / 원본주소 : {id(i)}, 복사본주소 : {id(copied_i)}")
```

```python
# 데이터 구조의 복제(x)
numbers = [5, 1, 9]   # 원본
new_numbers = numbers # 복사본x, 할당한 것
print(f"원본i={numbers}, 복수본i={new_numbers}) / 원본주소 : {id(numbers)}, 복사본주소 : {id(new_numbers)}")
new_numbers[0] = 555
print(f"원본i={numbers}, 복수본i={new_numbers}) / 원본주소 : {id(numbers)}, 복사본주소 : {id(new_numbers)}")
```

```python
# 복제(=깊은 복사)
numbers = [5, 1, 9]  # 원본
new_numbers = numbers.copy() # <- 복제본(깊은복사)
print(f"원본i={numbers}, 복수본i={new_numbers}) / 원본주소 : {id(numbers)}, 복사본주소 : {id(new_numbers)}")
new_numbers.sort()
print(f"원본i={numbers}, 복수본i={new_numbers}) / 원본주소 : {id(numbers)}, 복사본주소 : {id(new_numbers)}")
```
