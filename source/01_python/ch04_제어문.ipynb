# 1절. 조건문(if)
```
if 조건1:
    조건1이 true일 경우 수행할 명령어들(if문 단독 가능. 들여쓰기 중요)
elif 조건2:
     조건1이 거짓이고, 조건2가 참일 경우 수행할 명령어들
else:
      조건이 만족하지 않을 때 수행할 명령어들(else는 단독사용불가)
```

````python
score = int(input('Enter your score >'))
if score>=60:
    print("합격입니다")
else:
    print("불합격입니다")
````

## 2.1 for-in문
```
for 임시변수 in 나열가능한자료:
      반복할 문장들
else:
      나열 가능한 자료들 다 소요하여  for블록을 다 수행한 후, 최종 한번 실행할 문장들
      for문안의 break를 수행할 경우, else블록을 수행하지 않음(else는 선택)
```

````python
for ex1 in range(1, 10, 2): # [1, 3, 5, 7, 9]
    print(ex1, end=" ")
else:
    print("\nDONE")
    print('END')
````

````python
for ex1 in range(1, 10):
      if ex1==5:
            break # 현재 반복문을 빠져나오는 명령
            print(ex1, end=" ")
else:
    print('\nDONE')
print('END')
````

## 2.2 while문
```
while 조건:
                  조건이 참일 경우 반복할 문장들
else:
                  조건이 거짓일 경우 한번 실행할 문장(else 선택사항)
```

````python
# 1~100까지 짝수들의 합을 계산하고 출력하시오(while문)
tot = 0 # 누적변수
ex1 = 1
while ex1<=24: 
    if ex1%2==0: 
        tot += ex1 
    ex1 +=  1 # 이름1 변수 1증가 / 들여쓰기 추가적으로 더하면 입력값이 수행하지않음
else: # else를 입력하지않고 print하면 조건과 상관없이 모든 값이 출력됨.
    print('1~24까지 짝수의 합은', tot)
````

````python
# ex1 사용자에게 두 정수를 입력받아 작은 수부터 큰 수까지 홀수들의 합을 누적하여 출력하시오(pdf 9p.)
first = int(input('첫번째 정수는? '))
second = int(input('두번째 정수는? '))
# (1) 두 정수 중에 작은 수는 minNo, 큰 수는 maxNo에 할당
if first<second:
    minNo = first
    maxNo = second
else:
    minNo = second
    maxNo = first
minNo = first if first<second else second
maxNo = first if first>second else second
# (2) minNo~maxNo까지 홀수 누적하기
tot = 0
for no in range(minNo, maxNo+1):
    if no%2: # if no%2!=0
        tot += no
print(f"{minNo}부터 {maxNo}까지 홀수의 합은 {tot}입니다")
````

````python
# ex) 1~100사이의 정수를 맞추는 프로그램을 구현하시오. (프로그램안에 임의의 숫자가 만들어지고 그 숫자를 맞추도록 합니다)
'''
(프로그램 안에서 random수가 만들어지고 그 random수 맞추기)
1단계 : 수를 못 맞추면 UP/DOWN을 출력하고, 맞추면 축하메세지 출력 후 끝남 (1~100 사이의 수를 알아맞추기)
2단계 : 수를 못 맞추면 min~max를 조정, 맞추면 축하메세지 출력 후 끝남
3단계 : 2단계의 기회를 5회로 제한.  단, 5회의 기회가 지나면 실패메세지 출력 후 끝남
(2단계와 3단계는 연습문제로)
'''
import random
# min -> 해당 함수는 이미 있는 명령어로 해당 함수 등록X
rdN = random.randint(1, 100)
print('컴퓨터가 발생한 임의의 수 : ', rdN)
while True:
    inN = int(input("1~100 사이의 수를 맞춰 보세요 : "))
    if (inN>100) | (inN<1):
        print('Error. 재입력해주세요.')
    elif inN>rdN:
        print("Down ↓")
    elif inN<rdN:
        print("UP ↑")
    else:
        print('축하합니다! 맞추셨습니다!!')
        break
print('DONE')
````

````python
import random
# min -> 해당 함수는 이미 있는 명령어로 해당 함수 등록X
rdN = random.randint(1, 100)
print('컴퓨터가 발생한 임의의 수 : ', rdN)
inN = 0
while inN!=rdN:
    inN = int(input("1~100 사이의 수를 맞춰 보세요 : ")) # 2단계는 min~max사이의 수를 맞춰보세요
    if (inN>100) | (inN<1):
        print('Error. 재입력해주세요.')
    elif inN>rdN:
        print("Down ↓")
        # 2단계는 max값 조정
    elif inN<rdN:
        print("UP ↑")
        # 2단계는 min값 조정
    else:
        print('축하합니다! 맞추셨습니다!!')
print('DONE')
````

# 4절. 중첩루프 탈출

````python
for a in range(0,3): # a <= 0, 1, 2
    for b in range(1, 3): # b <=1, 2
        if a == b:
            print("a={}, b={} break".format(a, b))
            break
        print(a, b)
        
print("DONE")
````

````python
# 중첩 루프 탈출 방법1 : flag 변수 이용
flag = False
for a in range(0,3): # a <= 0, 1, 2
    for b in range(1, 3): # b <= 1, 2
        if a == b:
            print("a={}, b={} break".format(a, b))
            flag = True
            break
        print(a, b)
    if  flag:
        break
print("DONE")
````
