## 2.1 형식이 있는 덱스트 데이터 불러오기
- ch09_member.txt ->> 객체 list

````python
with open('data/ch09_member.txt', 'r', encoding='utf-8') as f:
    lines = f.readlines()
for line in lines:
    #print(line)
    data = line.strip().split(',')
    name = data[0]
    age = int(data[1].strip())
    email = data[2].strip()
    address = data[3].strip()
    print(f"이름:{name}, 나이:{age}, 메일:{email}, 주소:{address}")
````

````python
class Member:
    def __init__(self, name, age, email, address):
        self.name = name
        self.age = age
        self.email = email
        self.address = address
    def __str__(self):
        return "{},{},{},{}".format(self.name,
                                                self.age,
                                                self.email,
                                                self.address)
    def __eq__(self, other):
        if isinstance(other, Member):
            return self.__str__() == other.__str__()
        return False
    
    def as_dict(self):  # 객체를 딕셔너리로 return
        return {'name':self.name, 'age':self.age, 'email':self.email, 'address':self.address}
    def as_list(self): # 객체를 list로 return
        return [self.name, self.age, self.email, self.address]
````

````python
member = Member("홍길동", 33, 'hong@h.co.kr', '신림동')
print(member)
print(member.as_dict())
print(member.__dict__)
print(member.as_list())
````

# 4절. JSON파일에 저장하고 불러오기(dump, load)
## 4.1 JSON dump(파일에 저장)
1. 딕셔너리 리스트 데이터를 json파일로
2. 객체 리스트 데이터를 json파일로

````python
data = [{'name': '홍길동','age': '20','email': 'kildong@hong.com','address': '서울시 관악구'},
        {'name': '김길동','age': '40','email': 'kimdong@hong.com','address': '서울시 영등포구'},
        {'name': '신길동','age': '30','email': 'sindong@hong.com','address': '서울시 동작구'}]
````

````python
# 1. 딕셔너리 리스트를 json파일로 저장(dump)
# ensure_ascii옵션의 역할 : ASCII 코드 범위를 넘어선 문자를 어떻게 처리할지
# True(기본값) : ASCII코드 범위를 넘어선 한글은 유니코드로 저장
# False : 비ASCII 문자의 원래 형태로 저장
import json
with open('data/ch09_member1.json', 'w', encoding='utf-8') as jsonfile:
    json.dump(data, jsonfile, ensure_ascii=False, indent=' ')
````

````python
# 2. 객체리스트를 json파일로 저장(dump)
class Member:
    def __init__(self, name, age, email, address):
        self.name = name
        self.age  = age
        self.email= email
        self.address = address
    def as_dict(self):
        '객체를 딕셔너리로 바꿔 return (json.dump함수 사용에 필요)'
        return {"name":self.name, "age":self.age, "email":self.email, "address":self.address}
        # return self.__dict__
    def __str__(self):
        return "{}님 {}살 {}, {} 주거중".format(self.name, self.age, self.email, self.address)
    def __eq__(self, other):
        '''
        m1.__eq__(m2) 호출시 두 객체의 속성변수의 값이 모두 일치하는지 여부
        m1 == m2 구현시 m1.__eq__(m2)호출
        '''
        if isinstance(other, Member):
            return self.__str__() == other.__str__()
        else:
            return False
````

        * data/ch09_customers.txt 파일이 없으면 빈 파일 생성
- ch09_customers.txt 파일 내용은
```
홍길동, 010-9999-9999, k@hong.com, 30, 5, vvip
유길동, 010-5432-2222, y@hong.com, 20, 2, 가칠해
```
## Customer 클래스 작성
        * Customer("홍길동", "010-9999-9999", "k@hong.com", 30, 5, "vvip")
        * __init__, __str__, as_dict, to_txt_style(객체를 text 스타일 파일로 출력하기 위해)

````python
class Customer:
    def __init__(self, name, phone, email, age, grade, etc):
        self.name  = name
        self.phone = phone
        self.email = email
        self.age   = age
        self.grade = grade
        self.etc   = etc
    def as_dict(self):
        return self.__dict__
    def __str__(self): # ***  홍길동    010-9999-9999  k@k.com   30   vvip
        return "{:>5}\t{:3}\t{:15}\t{:15}\t{:3}\t{}".format(self.grade*'*',
                                                           self.name, 
                                                           self.phone,
                                                           self.email,
                                                           self.age,
                                                           self.etc)
    def to_txt_style(self): # 홍길동, 010-9999-9999, k@hong.com, 30, 5, vvip
        return "{}, {}, {}, {}, {}, {}".format(self.name, 
                                              self.phone,
                                              self.email,
                                              self.age,
                                              self.grade,
                                              self.etc)
````

````python
c1 = Customer("홍길동", "010-9999-9999", "h@h.com", 30, 4, "기타")
print(c1.as_dict())
print(c1)
print(c1.to_txt_style())
````

````python
def to_customer(row): # "홍길동, 010-9999-9999, k@hong.com, 30, 5, vvip\n"=>Customer객체
    data = row.strip().split(', ')
    name = data[0]
    phone = data[1]
    email = data[2]
    age   = int(data[3])
    grade = int(data[4])
    etc   = data[5]
    return Customer(name, phone, email, age, grade, etc)
````

````python
def fn1_insert_customer_info():
    '사용자로부터 고객정보를 입력받아 Customer형 객체를 return'
    import re
    name = input('이름 :')# abc, 123, 김
    name_pattern = r'[가-힣]{2,}'
    while not re.search(name_pattern, name):
        print('이름을 제대로 입력하세요(한글 2글자 이상)')
        name = input('이름 :')
    phone = input('전화 :')
    email = input('메일 :')
    while True:
        try:
            age = int(input('나이 :'))
            if (age<=0) | (age>130):
                raise Exception("나이 범위 이상")
            break
        except:
            print('올바른 나이를 입력하세요')
    while True:
        try:
                grade = int(input('등급(1~5) :'))
                if grade<1 :
                     grade = 1
  
                if grade > 5 :
                    grade = 5
        except:
                grade = 1
                print('유효하지 않는 등급 입력시 1등급으로 초기화')
        etc = input('기타 정보 :')
        return Customer(name, phone, email, age, grade, etc)
````

## 3. 종료

````python
def fn3_save_customer_txt(customer_list):
    'customer_list(객체list)의 내용을 ch09_customers.txt에 백업'
    customer_text_list = []
    for customer in customer_list:
        text = customer.to_txt_style()+'\n'
        customer_text_list.append(text)
    with open('data/ch09_customers.txt', 'w', encoding='utf-8') as f:
        f.writelines(customer_text_list)
    print("백업 종료합니다. 또 뵙겠습니다.")
````

````python
# 3번 테스트
fn3_save_customer_txt(customer_list)
````
