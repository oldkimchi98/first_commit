```python
def add(x, y=0):
    '''
    매개변수 2개를 받아 두자리 숫자로 만들어요
    x = 10의 자리수가 됨
    y = 1의 자리수가 됨(default 값 0)
    사용법 : add(2, 3) => 23
    '''
    return x*10 + y
```

```python
add(5,4)
```

```python
add(5)
```

```python
add(0,5)
```

```python
# 정규표현식(regex)

text = '''
   1. 참조: https://ko.wikipedia.org/wiki/정규_표현식
   2. 간략문법
       \d : (숫자와 매치. [0-9]와 동일)
       \D : (숫자가 아닌 문자)
       \w : (영문자나 숫자. [a-zA-Z0-9]와 동일)
       \W : (영문자나 숫자가 아닌 문자) [~!@#$%^&*()_\-+]
       .  : (문자)
       \. : (.)
       \s : ( )
       {n,} : n회 반복
       {2,4} : 2~4회 반복
       ? : (0번이나 1번 의미)
       + : (1번 이상 반복)
       * : (0번 이상 반복)
       () : 블록잡을 때
    3. 정규표현식 연습장 : https://regexr.com/ 
       ex1. 전화번호 패턴 : 0\d{1,2}[\-]\s\.]\d{3,4}[\-\s\.]\d{4}
       ex2. 이메일 패턴 : \w[a-z0-9_]+@\w+(\.\w+){1,2}
'''

print(text)
```

```python
data = "이름:홍길동, 주소:신림동, 전화:02-888-8888, 02-741-8475 02)888-7878\
핸드폰:010-9999-9999, 이메일:abc@ig.co.kr, abc@naver.com\
특징:서해번쩍 동해번쩍, 혈액형:B ㅋㅋ"
print(data)
```

```python
import re
tel_pattern = r"0\d{1,2}[\-)\s\.]\d{3,4}[\-\s\.]\d{4}"
tels = re.findall(tel_pattern, data) # 해당 패턴을 다 찾아줘. 없으면 빈 list([])
tels
```

```python
# 정규표현식 패턴과 처음 match되는 것을 찾아줘
firstTel = re.search(tel_pattern, data)
firstTel
```

```python
print(firstTel.group(0))
print(data[19:30])
```

```python
phone_pattern = r"(\d{2,3})-(\d{3,4})-(\d{4})"
phone = re.search(phone_pattern, data)
print(phone.group(0))
print(phone.group(1))
print(phone.group(2))
print(phone.group(3))
```
