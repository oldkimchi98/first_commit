# 2절. try ~ except로 예외처리
## 2.1 try~except
```
try:
        예외가 발생할 가능성이 있는 명령어들
except:
        예외가 발생했을 경우 실행할 명령어들
```

````python
# 100을 입력받은 정수값으로 나눠 출력
# ZeroDivisionError : division by zero ->> 예외 객체의 메세지
# ValueError: invalid literal for int() with base 10: 'a' ->> 잘못된 문자를 입력할 경우
x = int(input('정수를 입력하세요 > '))
print('입력한 정수는 {} '.format(x))
print('100을 입력한 정수로 나누면 {:.2f}'.format(100/x))
print('DONE')
````

````python
try:
    x = int(input('정수를 입력하세요 > '))
    print('입력한 정수는 {} '.format(x))
    print('100을 입력한 정수로 나누면 {:.2f}'.format(100/x))
except:
    print('나누기에 유효한 정수가 아닙니다')
print('DONE')
````

````python
# 100을 나눌 유효한 정수를 입력할 때까지 정수를 입력받아 유효한 정수를 입력하면 입력한 정수와 100으로 나눈 수를 출력
while True:
    try:
        x = int(input('정수를 입력하세요 > '))
        print('입력한 정수는 {} '.format(x))
        print('100을 입력한 정수로 나누면 {:.2f}'.format(100/x))
        break
    except:
        print('나누기에 유효한 정수가 아닙니다')
print('DONE')
````

## 2.2 예외를 지정한 처리
    try:
        예외가 발생할 수도 있는 문장들
    except 예외타입1:
        해당 예외가 발생할 경우 실행할 문장
    except 예외타입2:
        해당 예외가 발생할 경우 실행할 문장
        
여러 예외처리를 할 경우 상위 예외타입은 아래로

````python
# 100을 나눌 유효한 정수를 입력할 때까지 정수를 입력받아 유효한 정수를 입력하면 입력한 정수와 100으로 나눈 수를 출력
# ZeroDivisionError : "division by zero" 
# ValueError : invalid literal for int() with base 10: 'a'
while True:
    try:
        x = int(input('100을 나눌 정수를 입력하세요 >'))
        print('입력한 정수는 {}'.format(x))
        print('100을 입력한 정수로 나누면 {:.2f}'.format(100/x))
        break    
    except ValueError:
        print('정수가 아닙니다. 다시 입력하세요')
    except ZeroDivisionError:
        print('0으로 나눌 수 없습니다. 다시 입력하세요')    
    except Exception:
        print('원인 모를 예외가 발생했습니다. 다시 입력하세요')
print('DONE')
````

````python
# 100을 나눌 유효한 정수를 입력할 때까지 정수를 입력받아 유효한 정수를 입력하면 입력한 정수와 100으로 나눈 수를 출력
# ZeroDivisionError : "division by zero" 
# ValueError : invalid literal for int() with base 10: 'a'
while True:
    try:
        x = int(input('100을 나눌 정수를 입력하세요 >'))
        print('입력한 정수는 {}'.format(x))
        print('100을 입력한 정수로 나누면 {:.2f}'.format(100/x))
        break    
    except (ValueError, ZeroDivisionError):
        print('100으로 나눌 수 있는 정수가 아닙니다. 다시 입력하세요')   
    except Exception:
        print('원인 모를 예외가 발생했습니다. 다시 입력하세요')
print('DONE')
````

##    try:
    예외가 일어날 수도 있는 명령어들
    except [예외타입 [as e]]:
        해당 예외타입시 수행할 명령어
    [else:
        try블록을 다 수행한 후 실행할 명령어 ]
    finally:
        try블록이나 except블록을 수행한 후 실행

````python
try:
    f = open('data/ch08.txt', 'r') #encoding='utf-8') # FileNotFoundError
    data = f.read() # UnicodeDecodeError
except (FileNotFoundError, UnicodeDecodeError) as e:
    print(e)
else:
    print(data) # try블록을 다 실행한 후 실행
finally:
    f.close()
````

````python
# 사용자 정의 예외 : Exception클래스를 상속받은 클래스
class LengthZeroError(Exception):
    '매개변수 길이가 0일때 발생하는 예외'
    def __init__(self, message):
        super().__init__(message) # 부모클래스의 생성자에서 예외메세지를 셋팅
````

````python
def args_print(*data): # 튜플 매개변수
    if len(data)==0:
        raise LengthZeroError("매개변수 길이가 0이면 예외발생(사용자 정의 예외)")
    for item in data:
        print(item, end=' ')
    print('등을 입력하셨습니다')
````

````python
args_print('a','b','c')
````

````python
args_print()
````

````python
try:
    args_print()
except Exception as e:
    print(e)
````

# 4절. 추상클래스
- 추상클래스 : 추상메소드가 1개이상 포함된 객체를 생성할 수 없는 클래스
- 추상메소드 : 호출할 수 없는 메소드. 상속받은 클래스에서 재정의를 강요할 때
- 파이썬에서의 추상클래스는 raise를 이용하거나, @abstractmethod을 이용하여 추상메소드를 구현

````python
class Shape:
    def __init__(self):
            raise NotImplementedError('객체를 생성할 수 없음')
    def calc_area(self):
            raise NotImplementedError('추상메소드')
````

````python
from abc import abstractmethod, ABC
class Shape(ABC):
    @abstractmethod
    def __init__(self):
        pass
    @abstractmethod
    def calc_area(self):
        pass
````

````python
s = Shape()
````

````python
class Circle(Shape):
    def __init__(self, r=3):
        self.r = r
    def calc_area(self):
        import numpy as np
        return np.pi * (self.r**2)
````

````python
myCircle = Circle(5)
myCircle.calc_area()
````
