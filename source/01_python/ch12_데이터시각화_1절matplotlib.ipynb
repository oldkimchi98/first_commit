```python
import numpy as np
x = np.linspace(0, 9.9, 100)
x.shape
```

```python
# plt.figure(figsize=(5,5)) # 그래픽 사이즈가 해당 셀만 적용
plt.rcParams['figure.figsize'] = (4,4) # 그래프 사이즈 셋팅(다음셀도 적용)
plt.subplot(2,1,1) # 2행 1열의 subplot에서 1번째 subplot를 다음과 같이 그린다
plt.plot(x, np.sin(x))
plt.subplot(2,2,3) # 2행 2열의 subplot에서 3번째 subplot를 다음과 같이 그린다
plt.plot(x, np.cos(x))
# plt.subplot(2,2,4)
plt.subplot(224)
plt.plot(x, np.sin(x)*np.cos(x))
plt.show()
```

### 2) subplots()함수로 서브플롯 생성
- plot함수에서 색상이나 스타일을 조정 : https://wikidocs.net/92085

```python
plt.subplots(nrows=2, ncols=2, figsize=(7,4))
```

```python
# plt.plot(x, y, color='limegreen')
# plt.plot(x, y, marker="*")
# plt.plot(x, y, '-.')
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(7,4))
axes[0,0].plot(x,np.sin(x), color='limegreen', linestyle='-.') # color='r', linewidth=10
axes[0,1].plot(x, np.cos(x), 'c--') # color='b', linestyle='--'(대시라인), :(점선)
axes[1,0].plot(x, np.tanh(x), '#00FF00') # c='#00FF00'
axes[1,0].plot([0,2,4],[-1,0,1], 'k*-.') # c='k',marker='o', linestyle='-.'
axes[1,1].plot(x, np.sin(x)*np.cos(x), 'r--.')
```

```python
# 1행4열 서브플롯 (for문)
x = np.linspace(0, 9.9, 10)
func_list = [np.sin, np.cos, np.tanh, lambda x:np.sin(x)*np.cos(x)]
fmt_list = ['co--', 'y*--', '#00AA00', 'mX:']
fig, axes = plt.subplots(1,4, figsize=(9,2), sharey=True) # y축 공유
for i, ax in enumerate(axes):
    ax.plot(x, func_list[i](x), fmt_list[i])
axes[2].plot([3,6],[-1, 1])
```

### 4) 축과 눈금
- https://matplotlib.org/3.1.0/api/axis_api.html (교안 33p)

```python
import numpy as np
x = np.linspace(0, 10, 100)
y = np.cos(x)

fig, axes = plt.subplots(2,2, figsize=(8,5))
plt.subplots_adjust(hspace=0.4) # hspace위아래여백(1은 서브플롯세로높이)
axes[0,0].scatter(x,y, c='orangered', ls='-.')
axes[0,0].set(title='An example', ylabel='y값', xlabel='x값', ylim=[-2,2], xlim=[3,6], yticks=[-2, -1, 0, 1, 2], yticklabels = ['limit_min', -1, '원점', 1, 'limit_max'],
                        xticks=[3,4,5,6], xticklabels=['삼','사점',5,'6점'])
# ylim=[y축 눈금 범위], yticks=[y축 눈금], yticklabels = [y축 눈금의 라벨]
axes[0,1].scatter(x,y, c='m', marker='X')
axes[0,1].set_title('두번째', fontsize=15)
axes[0,1].set_xticks(range(0,11,3))
axes[0,1].set_ylim(-1.5, 1.5)

axes[1,0].scatter(x,y, c='violet')
axes[1,0].spines['top'].set_visible(False)
axes[1,0].spines['bottom'].set_position( ('outward', 10)) # 10pt만큼 바깥으로 이동

axes[1,1].bar(x,y, color='limegreen')
axes[1,1].grid(True)
plt.show()
```

### 5) ★ 축공유 ★

```python
import numpy as np
x = np.arange(0,10)
y1 = 0.5 * x**2
y2 = -1 * y1
```

```python
fig, ax1 = plt.subplots(figsize=(6,3))
ax1.plot(x,y1, 'm^:')
ax1.set(title='축공유', xlabel='X data', ylabel='y1 data')

ax2 = ax1.twinx() # ax1과 x축을 공유하는 ax2
ax2.plot(x,y2,'b--')
ax2.set_ylabel('y2 data', color='b')

ax3 = ax1.twiny() # ax1과 y축을 공유하는 ax3
ax3.plot(-x, y1, 'rv-')
ax3.set_xlabel('-x data', color='r')
plt.show()
```
