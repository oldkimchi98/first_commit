````python
data = "이름:홍길동, 주소:신림동, 전화:02-888-8888, 02-741-8475 02)888-7878\
핸드폰:010-9999-9999, 이메일:abc@ig.co.kr, abc@naver.com\
특징:서해번쩍 동해번쩍, 혈액형:B ㅋㅋ"
print(data)
````

````python
import re
tel_pattern = r"0\d{1,2}[\-)\s\.]\d{3,4}[\-\s\.]\d{4}"
tels = re.findall(tel_pattern, data) # 해당 패턴을 다 찾아줘. 없으면 빈 list([])
tels
````

````python
# 정규표현식 패턴과 처음 match되는 것을 찾아줘
firstTel = re.search(tel_pattern, data)
firstTel
````

````python
print(firstTel.group(0))
print(data[19:30])
````

````python
phone_pattern = r"(\d{2,3})-(\d{3,4})-(\d{4})"
phone = re.search(phone_pattern, data)
print(phone.group(0))
print(phone.group(1))
print(phone.group(2))
print(phone.group(3))
````

````python
email_pattern = r"\w[a-z0-9_]+@\w+(?:\.[a-z]+){1,2}"
print(re.findall(email_pattern, data))
print(re.search(email_pattern, data). group(0))
````

````python
# 특정 문자외 없애기
re.sub(r"[^가-힣0-9\-a-z@]", " ", data)
````

````python
# %s = 문자, %d = 10진수정수, %f = 실수
print("%s님 나이는 %d살입니다"%(name, age))
print(f"{name}님 나이는 {age}살입니다")
````

````python
print("{}님 나이는 {}살입니다".format(name, age))
print("{0}님 나이는 {1}살입니다".format(name, age))
print("{1}님 나이는 {0}살입니다".format(age, name))
print("{0:4.1f}살 {1:@>4.2}님 반갑습니다. {1}님 어서오세요".format(age,name))
````

## 4.2 format()문법

- "{[인덱스]:[공백대체문자].[정렬방식].[자릿수].[타입]}".format(변수명)
    * 인덱스 : format 함수 안의 매개변수 순서
    * 공백대체문자는 정렬방식을 명시하여야 함. 자릿수를 여유롭게
    * 정렬방식 : <(왼쪽정렬), >(오른쪽정렬), ^(가운데정렬)
    * 정수자릿수 : 출력할 자리수
    * 실수자릿수 : 전체출력될자리수. 변수 중 출력할 자리수
    * 타입 : d(10진수), f(실수4.1f은 소수점자리수), s(문자), o(8진수), x(16진수), b(2진수)

````python
name = input("이름은?")
age = int(input("나이?"))
````

````python
print("[{0:~>10}]살 [{1:*^10}]님 안녕하세요".format(age, name))
print(f"[{age:^<10.1f}]살 [{name:~>10.2}]님 안녕하세요")
````

````python
a = 2.755
sign = '+'
width = 10
digits = 3
print("[{:+10.3}]".format(a))
print("[{:{sign}{width}.{digits}}]".format(a,
                                           sign=sign,
                                           width=width,
                                           digits=digits))
````

### 4.3 날짜출력

````python
# Y(2026) y(26) m(월) d(일) H(24시) M(분) S(초) I(12시) p(am,pm)
print("{:%Y년%m월%d일 %H시%M분%S초}".format(day))
print("{:%Y년%m월%d일 %p %I시%M분%S초}".format(day))
print(day.strftime("%Y년%m월%d일 %p %I시%M분%S초"))
# 날짜를 문자열로 전환하는 함수 : 날짜형변수.strftime(포맷)
````
