# 2. 주피터 노트북 실행법
- shift+Enter : 포커스된 셀이 실행되고 커서가 다음셀로 이동
- alt+Enter : 포커스된 셀이 실행되고 다음셀이 생성된 후 다음셀로 이동
- ctrl+Enter : 포커스된 셀이 진행되고 포커스 이동이 없음
- Enter : 다시 편집 상태로
