## 5.3 논리연산자 : and(&), or, |, not

```python
# and : 논리연산자, 거짓으로 판별되는 첫번째 항의 결과 반환
# 모든 항이 참이면 마지막항을 반환
print( False and True and 'abc' )
print(0 and 5)
```

```python
# & : 논리연산자, 비트연산자 
# & > 비교연산이 and보다 우선순위 > and
print( (10>3) | (10>5) ) # 논리연산 True | True => True
print( 3 | 10 ) # 비트연산 수&수
#  3: 0 0 1 1
# 10: 1 0 1 0
# | : 1 0 1 1 (11)
print( 10>3 | 10>5) # 10>11>5
```

```python
# or : 논리연산자, 참으로 판별되는 첫번째 항의 결과 반환
# 모든 항이 거짓이면 마지막항을 반환
print( False or False or 0 )
print( 0 or '홍길동')
```

```python
not 10>5
```

# 6절 문자열 다루기 (명령어 모음)
- 문자열 슬라싱변수[from:stop:step]
- len(문자변수) : 문자수 (확인용)
- upper() : 대문자로
- lower() : 소문자로
- title() : 각 어절의 앞글자만 대문자로
- capitalize() : 첫문자만 대문자로
- count('찾을 문자열') : '찾을문자열'이 몇개 들어있는지 알려줌. 없으면 0
- find('찾을문자열') : '찾을문자열'이 왼쪽부터 시작하여 몇번째에 있는지 index. 없으면 -1
- rfind('찾을문자열') : '찾을문자열'이 오른쪽부터 시작하여 몇번째에 있는지 index. 없으면 -1
     cf. index('찾을문자열') : '찾을문자열'이 왼쪽부터 시작하여 몇번째에 있는지 index. 없으면 error (잘안씀.)
- startswith('특정문자열') : '특정문자열'로 시작하는지 여부 
- endswith('특정문자열') : '특정문자열'로 끝나는지 여부

- strip() : 좌우공백제거(다른 언어로는 trim)
- rstrip() : 우측 공백제거
- lstrip() : 좌측 공백제거
- split(나눌문자) : 공백이나 다른 문자를 기준으로 나누어 리스트로 반환
- isdigit() : 문자열이 전부 숫자 문자열인지 여부
- islower() : 문자열이 전부 소문자인지 여부
- isupper() : 문자열이 전부 대문자인지 여부
- replace(oldStr, newStr) : 문자열안의 oldStr을 newStr로 변경한 값을 반환

```python
temp = "python is easy. pyTHon is smart. python"
     #  012345678901234567890123456789012345678
print('대문자:', temp.upper())
print('소문자:', temp.lower())
print('타이틀: ', temp.title())
print('capitalize :', temp.capitalize())
print('python이 몇번 나오는지? :', temp.count('python'))
print('대소문자 구분없이 python이 몇번 나오는지? :', temp.upper().count('PYTHON'))
print('대소문자 구분없이 python이 몇번 나오는지? :', temp.lower().count('python'))
print('대소문자 구분없이 python이 몇번 나오는지? :', temp.title().count('Python'))
print('python이 나오는 첫번째 index 위치 :', temp.find('python'))
print('java가 나오는 첫번째 index위치 :', temp.find('java'))
print('python이 마지막으로 나오는 index위치 :', temp.rfind('python'))
print('Python이 대문자로 시작하는지 여부 :', temp.startswith('Python'))
print('n으로 끝나는지 여부 :', temp.endswith('n'))
print('python대신 R로 변경 :', temp.replace('python', 'R').replace('pyTHon', 'R')) # 매우 자주 사용하므로 중요
print(temp)
```
